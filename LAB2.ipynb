```python
from collections import deque

def bfs(graph, start):

  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      visited.add(vertex)
      print(vertex)  # Print the visited vertex
      queue.extend(neighbor for neighbor in graph[vertex] if neighbor not in visited)

  return visited

graph = {
  'A': ['B', 'C'],
  'B': ['D', 'E'],
  'C': ['F'],
  'D': [],
  'E': ['F'],
  'F': []
}

start_vertex = 'A'
visited_vertices = bfs(graph, start_vertex)
print("Visited vertices:", visited_vertices)
```

```python
import numpy as np
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
print(type(list1))
print(type(list2))

array1 = np.array(list1)
array2 = np.array(list2)
print(type(array1))
print(type(array2))

multiplication  = array1*array2
print("multiplication",multiplication)

division = array1/array2
print("division",division)

poweroff = np.power(array1, array2)
print("poweroff", poweroff)


addition = array1 + array2
print("addition",addition)

sine = np.sin(array1)
print("Sine:", sine)

log = np.log(array1)
print("Logarithm (base e):", log)

log2 = np.log2(array1)
print("Logarithm (base 2):", log2)

exp = np.exp(array1)
print("Exponential:", exp)
```

```python
import matplotlib.pyplot as plt

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(age, height, color='blue', marker='o')
plt.title('Relationship between Age and Height')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.bar(people, weight, color='green')
plt.title('Weight of Individuals')
plt.xlabel('Person')
plt.ylabel('Weight (kg)')


plt.subplot(1, 3, 3)
plt.hist(height, bins=4, color='purple', edgecolor='black')
plt.title('Distribution of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```
